```python
# checking unique values of the columns
unique_data =  df['Country Name'].unique()
print(unique_data)
print("Total number of unique countries:" ,df['Country Name'].nunique())
```

```python
# Extracting top 10 country with respect to total population
total_population = df[df['Series Code'] == 'SP.POP.TOTL']
# sort the total population with highest based on 2022
total_population_sorted = total_population.sort_values(by = '2022',ascending = False)
total_population_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries based on 2022 population\n")
print(total_population_top_ten_countries[['Country Code']])
```

```python
# ploting world population using histogram
plt.figure(figsize=(15, 6))
plt.subplot(2, 2, 1)
sns.barplot(x = "Country Name",y = "2022",hue = "Country Name",data = total_population_top_ten_countries ,palette='coolwarm',legend = False)
plt.title("Distribution of Total Population (2022)", fontsize=10)
plt.xlabel("Country", fontsize=10)
plt.ylabel("Total Population", fontsize=10)
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
